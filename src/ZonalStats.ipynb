## Importations

```python
# Importations
import pandas as pd
import os 
import numpy as np
import glob
```

# Data

```python
# Load datasets
BASE_PATH = os.path.join('..', 'results', 'attention_maps_l19')
EXCEL_FILES = glob.glob(os.path.join(BASE_PATH, '*0.xlsx'))
```

```python
# Populate DF
DFs = {}
for x in EXCEL_FILES:
    DFs[x.split(os.path.sep)[-1][:-5]] = pd.read_excel(x)
DF = pd.concat([df for df in DFs.values()], axis=0)
print("DF dimensions:", DF.shape)
DF.head()
```

```python
# Original ID
def splitID(row):
    return int(row[2:])
DF['ID'] = DF['ID'].apply(splitID)
```

```python
# Sort
DF = DF.sort_values('ID')
DF = DF.reset_index()
DF.drop(columns=['index'], inplace=True)
DF
```

```python
# Load class prob
DFClass = pd.read_excel(os.path.join(BASE_PATH, 'ClassProbs.xlsx'), index=None)
DFClass = DFClass.sort_values('ID')
DFClass = DFClass.reset_index()
DFClass.drop(columns=['index', 'Unnamed: 0'], inplace=True)
DFClass
```

```python
# Merge
DFFull = pd.merge(DF, DFClass, how='left', on='ID')
print("Consolidated dimension:", DFFull.shape)
DFFull
```

# Extra stats 
Specific thresholds based on land cover types

## High risk
## lcV3" >=0.35 AND "lcV8" >=0.1

```python
# HR
HR = DFFull[(DFFull['lcV3_30%'] >= 0.35) & (DFFull['lcV8_30%'] >= 0.1)]
HR.shape
```

```python
# Importations for evaluation
from sklearn.metrics import (mean_squared_error, roc_auc_score, auc, 
                             confusion_matrix, accuracy_score, classification_report,
                             precision_score, recall_score, f1_score, cohen_kappa_score)

# Reference
ytest = HR.y
y_hat = HR.yhat
y_score = HR.ProbF

# Evaluation
print("Test accuracy [%]:", np.round(accuracy_score(y_true=ytest, y_pred=y_hat) * 100, 2))
print("Precision Score:", np.round(precision_score(y_true=ytest, y_pred=y_hat),3))
print("Recall score:", np.round(recall_score(y_true=ytest, y_pred=y_hat), 3))
print("F1 score:", np.round(f1_score(y_true=ytest, y_pred=y_hat), 3))
print("Cohen Kappa score:", np.round(cohen_kappa_score(ytest, y_hat),3))
print("ROC AUC score:", np.round(roc_auc_score(y_true=ytest, y_score=y_score), 3))
tn, fp, fn, tp = confusion_matrix(ytest, y_hat).ravel()
print("Specificity:", tn / (tn + fp))
print("Sensitivity:", tp / (tp + fn))
print(confusion_matrix(ytest, y_hat))
```
